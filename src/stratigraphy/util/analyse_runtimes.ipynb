```python
import pydantic
import abc
import fitz

class Rectangle(pydantic.BaseModel):
    x0: float
    y0: float
    x1: float
    y1: float

    @property
    def width(self):
        return self.x1 - self.x0

    @property
    def height(self):
        return self.y1 - self.y0
    

class Coordinate(pydantic.BaseModel):
    east: float
    north: float
    coordinate_system: str = 'LV03'
    rectangle: Rectangle = None



# Create an instance of Coordinate with keyword arguments for the Rectangle
coordinate = Coordinate(
    east=240.1,
    north=300.2,
    coordinate_system='LV95',
    rectangle=Rectangle(x0=0.0, y0=0.0, x1=100.0, y1=100.0)
)
coordinate.model_dump()

```

```python
class Text(pydantic.BaseModel):
    text: str
    rectangle: Rectangle

    @property
    def area(self):
        return self.rectangle.width * self.rectangle.height
    

text = Text(text='Hello, World!', rectangle=Rectangle(x0=0.0, y0=0.0, x1=100.0, y1=100.0))
text.model_dump()
```

```python
class Number(pydantic.BaseModel):
    value: float
    rectangle: Rectangle

    @property
    def area(self):
        return self.rectangle.width * self.rectangle.height
    
number = Number(value=42.0, rectangle=Rectangle(x0=0.0, y0=0.0, x1=100.0, y1=100.0))
number.model_dump()
```
